```python
import matplotlib.pyplot as plt
from matplotlib import image as mpimg
```

```python
plt.figure(figsize=(10, 10))
f,ax=plt.subplots(1,2)
image2 = mpimg.imread("images/BD_Baidu_08_dehazed.jpg")
image1 = mpimg.imread("images/BD_Baidu_08.png")
ax[0].imshow(image1)
ax[0].set_title("CA : hazed image")
ax[1].imshow(image2)
ax[1].set_title("CA : dehazed image")
plt.show()
```
